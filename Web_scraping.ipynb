```python

import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    """
    Function to plot stock data.
    :param data: DataFrame containing stock data with a 'Date' column and a 'Close' column
    :param title: Title for the graph
    """
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

tesla_data = yf.download("TSLA", start="2010-01-01", end="2025-01-01")

tesla_data.reset_index(inplace=True)

make_graph(tesla_data, "Tesla Stock Price Over Time")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):

    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price', color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

gme_data = yf.download("GME", start="2010-01-01", end="2025-01-01")

gme_data.reset_index(inplace=True)

make_graph(gme_data, "GameStop Stock Price Over Time")
```
